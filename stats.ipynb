```python
import matplotlib.pyplot as plt

# Data setup
thread_counts = ['Serial', '1', '2', '4', '8', '16']
data_sizes = ['10000', '100000', '500000', '1000000']
```

```python
# Time data for each configuration
insert_time_serial = [0.019, 0.168, 1.002, 4.235]
insert_time_1 = [0.022, 0.208, 1.132, 6.612]
insert_time_2 = [0.023, 0.208, 1.023, 4.952]
insert_time_4 = [0.018, 0.141, 0.868, 3.323]
insert_time_8 = [0.020, 0.124, 0.603, 2.089]
insert_time_16 = [0.031, 0.229, 1.034, 4.245]

# Combining the data
times = [insert_time_serial, insert_time_1, insert_time_2, insert_time_4, insert_time_8, insert_time_16]

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
for idx, time in enumerate(times):
    ax.plot(data_sizes, time, label=f'{thread_counts[idx]} Threads', marker='o')
    
ax.set_xlabel('Data Size')
ax.set_ylabel('Time (seconds)')
ax.set_title('Insertion Times by Thread Count and Data Size')
ax.legend()

plt.show()
```

```python
search_time_serial = [0.023, 0.231, 1.666, 4.235]
search_time_1 = [0.022, 0.241, 1.776, 4.336]
search_time_2 = [0.023, 0.216, 1.520, 3.136]
search_time_4 = [0.018, 0.166, 0.945, 2.050]
search_time_8 = [0.022, 0.146, 0.668, 1.362]
search_time_16 = [0.031, 0.229, 1.094, 2.703]

# Combine all times into a list for plotting
search_times = [search_time_serial, search_time_1, search_time_2, search_time_4, search_time_8, search_time_16]

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
# Plotting each thread count's data
for index, times in enumerate(search_times):
    ax.plot(data_sizes, times, marker='o', label=f'{thread_counts[index]} Threads')

# Set labels and title for the plot
ax.set_xlabel('Data Size')
ax.set_ylabel('Search Time (seconds)')
ax.set_title('Search Times by Thread Count and Data Size')
ax.legend(title="Thread Counts")

# Display the plot
plt.show()
```
